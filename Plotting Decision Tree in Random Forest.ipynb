```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
# try this
# df = pd.DataFrame([iris])
# df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 30,random_state=30)
```

```python
X = iris.data
y = iris.target
model = model.fit(X,y)
```

```python
# To visualise all 30 trees
model.estimators_
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model.estimators_[1],filled = True)

# Note: Here many DT will be created and "estimators_" is a list containing info of all the 30 trees
# So we will have to mention the tree which we want to visualise by "model.estimators_[1]" i.e for 1st
# Change the value [1] to check other DT
```

```python
# for displaying all the DT we can write a function

for i in range(len(model.estimators_)):
    plt.figure(figsize=(15,10))
    tree.plot_tree(model.estimators_[i],filled = True)
    
```

```python
# For checking the detailed information of all the trees

for i in range(len(model.estimators_)):
    print("---------------------------------Decision Tree {}----------------------------------".format(i))
    print(tree.export_text(model.estimators_[i]))
```
